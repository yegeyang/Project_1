```python
import pandas as pd
import numpy as np
```

```python
file ="towed-vehicles.csv"
```

```python
vehicles_df = pd.read_csv(file)
vehicles_df.head()
```

```python
vehicles_df["Tow Date"]=vehicles_df["Tow Date"].map(lambda x: x.rstrip('T00:00:00.000'))
```

```python
vehicles_df.head()
```

```python
del vehicles_df["Model"]
vehicles_df
```

```python
vehicles_df.count()
```

```python
vehicles_df = vehicles_df.dropna(subset =["State"])
vehicles_df.count()
```

```python
vehicles_df
```

```python
vehicles_df["Make"].value_counts()
```

```python
vehicles_df["Make"] = vehicles_df["Make"].replace({'JAG':'JAGU','INTL':'INFI','TOYO':'TOYT','VOLK':'VW','MERZ':'MERC','LNDR':'LAND','GM':'GMC','SUZU':'SUZI'})
```
